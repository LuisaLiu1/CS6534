```python
def cal_gene_correlation(cell_gene_m):
    gene_gene_corr_matrix = cell_gene_m.corr()
    return gene_gene_corr_matrix

def cal_knn_matrix(K,conformation_m):
    distance_matrix = euclidean_distances(conformation_m)
        # 计算K最近邻
    nbrs = NearestNeighbors(n_neighbors=K, metric='precomputed').fit(distance_matrix)
    distances, indices = nbrs.kneighbors(distance_matrix)
    # 构建KNN邻接矩阵
    knn_adjacency_matrix = np.zeros_like(distance_matrix)
    for i in range(distance_matrix.shape[0]):
        knn_adjacency_matrix[i, indices[i, 1:]] = 1 # 注意排除自身
        knn_adjacency_matrix[indices[i,1:],i]=1

    # 保存结果
    filename='Knn_adja/knn_'+str(K)+'_adjacency_matrix.csv'
    np.savetxt(filename, knn_adjacency_matrix, delimiter=',')
    return knn_adjacency_matrix

def cal_pre_matrix(cor_m,knn_m,K,lam):
    model = GraphicalLassoCV(alphas=lam)
    model.fit(cor_m)
    precision_matrix = model.precision_
    # 添加约束矩阵A
    precision_matrix[knn_m == 0] = 0
    file="pre_matrix/presicion_matrix_k="+str(K)+"aplha="+str(lam)+".csv"
    np.savetxt(file, precision_matrix,delimiter=',')
    return precision_matrix

def get_top_n_entries(n,pre_m,K,lam):
    # 找出精度矩阵中前10大元素的索引(仅取上三角部分)
    triu_indices = np.triu_indices(pre_m.shape[0], 1)
    top_indices = np.argsort(np.abs(pre_m[triu_indices]))[-n:]
    row_indices = triu_indices[0][top_indices]
    col_indices = triu_indices[1][top_indices]

    files="top_entries/top_"+str(n)+"k="+str(K)+"aplha="+str(lam)+".txt"

    with open(files, "w") as f:
        for i, j in zip(row_indices, col_indices):
            f.write(f"({i}, {j}): {pre_m[i, j]}\n")

    # 打印出对应的基因对及其精度值
    for i, j in zip(row_indices, col_indices):
        print(f"({i}, {j}): {pre_m[i, j]}")

def main():
    K=range(5,55,5)
    lam = range(1, 13, 3)
    cor_matrix=cal_gene_correlation(cell_gene_m)
    for k in K:
        print("---K=",k)
        knn_m=cal_knn_matrix(k,conformation_m)
        for l in lam:
            print("------alpha=",l)
            pre_m=cal_pre_matrix(cor_matrix,knn_m,k,l)
            get_top_n_entries(10,pre_m,k,l)


if __name__ == '__main__':
    main()

    
```
